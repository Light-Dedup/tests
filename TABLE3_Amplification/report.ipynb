```python
import pandas as pd

with open("./table-calculated", "r") as f:
    df = pd.read_csv(f, delim_whitespace=True, engine='python')
df.head()
```

- Read/Write Amplification

```python
first = df.iloc[0]
second = df.iloc[1]

WA = first["write"] / 40
RA = second["read"] / 40

print("WA: " + str(first["write"]) + "/40", WA)
print("RA: " + str(second["read"]) + "/40", RA)
```
